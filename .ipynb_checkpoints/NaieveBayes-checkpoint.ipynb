This project aims to investigate if sleep and lifestyle factors can \

This notebook uses Gaussian Naieve Bayes to try and predict the presence of sleeping disorders given quality of sleep quality values and stress level values. Sleep Quality is a self reported rating, ranging from 1-10 and Stress Level is also a self reported rating, ranging from 1-10.

Standard Imports

```python
%matplotlib inline
import pandas as pd
import matplotlib. pyplot as plt
import matplotlib.patches as mp
plt.style.use("seaborn-v0_8-whitegrid")
import numpy as np

from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
dataset = pd.read_csv("Sleep_Health_and_Lifestyle_Dataset.csv")
```

This first graph just explores and visualizes the data to find any obvious correlations. It is expected that individuals with lower sleep quality and higher stress will be more likely to have insomnia and sleep apnea.

```python
# Map Sleep Disorder to colors
colour_map = {"Insomnia":"red", "Sleep Apnea":"blue"}
colours = dataset["Sleep Disorder"].map(colour_map).fillna("green")

plt.ylim(0,10)
plt.xlim(0, 10)
# Added opacity to help with visualization
plt.scatter(dataset["Stress Level"], dataset["Quality of Sleep"], c=colours, alpha=0.3, s=20)
plt.xlabel("Stress Level")
plt.ylabel("Quality of Sleep")
plt.title("Naieve Bayes: Stress Level vs Sleep Quality")

# Added legend
patches = [
    mp.Patch(color='green', label='None'),
    mp.Patch(color='red', label='Insomnia'),
    mp.Patch(color='blue', label='Sleep Apnea')
]
plt.legend(handles=patches, title='Sleep Disorder')

# Print correlation coefficients
print("Correlation Coefficients for Insomnia")
print(dataset[["Stress Level", "Quality of Sleep"]].corrwith((dataset['Sleep Disorder'] == 'Insomnia').astype(int)))
print("Correlation Coefficients for Sleep Apnea")
print(dataset[["Stress Level", "Quality of Sleep"]].corrwith((dataset['Sleep Disorder'] == 'Sleep Apnea').astype(int)))
```

From looking at this graph we can see that sleep apnea appears across a wide range of stress levels and sleep quality levels. This means there is not a strong correalation between these features and sleep apnea. However there may be a slight correlation between the presence of insomnia and sleep quality and stress. In this model insomnia only appears when sleep quality is less than 8 and stress is higher than 3. \
A stress level correlation coefficient of 0.13 for Insomnia indicates a small positive correalation between stress and the presence of insomnia in an individual. The quality of sleep correlation coefficient for insomnia of -0.33 indicates a negative correlation between quality of sleep and the presence of insomnia in an individual. A stress level correlation coefficient  of 0.082 for sleep apnea indicates a very weak positive correlation between stress level and the presence of Sleep Apnea in an individual. A Quality of sleep  correlation coefficient of -0.046 for sleep apnea indicates a very weak negative correlation between quality of sleep and the presence of sleep apnea in an individual.

Gaussian Naieve Bayes

The below Gaussian Naieve Bayes model aims to classify different sleep disorders and investigate the relationship between Quality of Sleep, Stress Level and Sleep Disorders. It investigates if it is possible to predict the presence of a specific sleep disorders (Insomnia and Sleep Apnea) given stress and sleep quality levels.  An instance where an individual does not have a sleep disorder appears as a green dot on the graph, insomnia appears as a red dot and sleep apnea appears in blue. In this first example a Gaussian Naieve Bayes model is trained using the dataset. Then 2000 random points are generated to see how the model classifies them. 

```python
# Fill any nan values with "None"
dataset["Sleep Disorder"] = dataset["Sleep Disorder"].fillna("None")

# Get features and column stack to get coordinates
stress = dataset["Stress Level"].values
sleepQuality = dataset["Quality of Sleep"].values
coords = np.column_stack((stress, sleepQuality))

# Encode labels
le = LabelEncoder()
labels = le.fit_transform(dataset["Sleep Disorder"])

# Train Gaussian model 
model = GaussianNB()
model.fit(coords, labels)

# Generate 2000 random points and predict which disorder each will have
rng = np.random.RandomState(0)
Xnew = [0,0] + [10,10] * rng.rand(2000, 2)
ynew = model.predict(Xnew)

# Mapping colours for graph
colour_map_nb = {cls: col for cls, col in zip(le.classes_, ["red", "green", "blue"])}
colours = [colour_map_nb[cls] for cls in le.inverse_transform(labels)]
ynew_colours = [colour_map_nb[cls] for cls in le.inverse_transform(ynew)]

# Plot the original points and then plot the random points
plt.ylim(0,10)
plt.xlim(0,10)
plt.scatter(coords[:, 0], coords[:, 1], c=colours, s=50, alpha=0.5)
plt.scatter(Xnew[:, 0], Xnew[:, 1], c=ynew_colours, s=20, alpha=0.1)

# Add legend, labels and title
patches = [
    mp.Patch(color='green', label='None'),
    mp.Patch(color='red', label='Insomnia'),
    mp.Patch(color='blue', label='Sleep Apnea')
]
plt.legend(handles=patches, title='Sleep Disorder')
plt.xlabel("Stress Level")
plt.ylabel("Quality of Sleep")
plt.title("Naive Bayes: Stress Level vs Sleep Quality")
plt.show()
```

Here we can see how the model classifies different areas of the graph. Some areas of the graph are likely inaccurate such as from 6-10 in quality of sleep and 0-2 in stress level, this is likely due to a lack of data in this area. There also appears to be a lot of overlap between the insomnia class and the none class.

The below Gaussian Naieve Baye implementation splits the dataset into training and test subsets in order to train the Gaussian Naieve Baye model and evaluate its performance.

```python
import seaborn as sns

# random_state ensures reproducibility 
# Data is split into training and testing samples, 30% of the data is used for testing and 70% is used for training
X_train, X_test, y_train, y_test = train_test_split(coords, labels, test_size=0.3, random_state=42)

# Train model
model = GaussianNB()
model.fit(X_train, y_train)

# Use model to predict test set
y_pred = model.predict(X_test)  

plt.ylim(0,10)
plt.xlim(0, 10)

# Map training colours, and test colours
true_colours = [colour_map_nb[cls] for cls in le.inverse_transform(y_test)]
pred_colours = [colour_map_nb[cls] for cls in le.inverse_transform(y_pred)]

# Test set points = circles
plt.scatter(X_test[:,0], X_test[:,1], c=true_colours, s=60, edgecolor='k', label='True Labels (Circles)')

# Predicted points = triangles
plt.scatter(X_test[:,0], X_test[:,1], c=pred_colours, s=30, marker='^', edgecolor='k', label='Predicted Labels (Triangles)')

# Add legend, labels and title
patches = [
    mp.Patch(color='green', label='None'),
    mp.Patch(color='red', label='Insomnia'),
    mp.Patch(color='blue', label='Sleep Apnea')
]
plt.legend(handles=patches, title='Sleep Disorder')
plt.xlabel("Stress Level")
plt.ylabel("Quality of Sleep")
plt.title("Naive Bayes: Stress Level vs Sleep Quality")
plt.show()
```

In the above graph, circles correspond to points from the test set and triangles correspond to points predicted by the model. As we can see the model does not always predict correctly, especially in the area where insomnia and no disorder points overlapped in the last graph. The model appears to be most effective for predicting when an individual does not have any sleeping disorders, but struggles to correctly predict cases of insomnia.

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")

# Heat map
sns.heatmap(cm.T, square=True, annot=True, fmt='d', cbar=False,
            xticklabels=le.classes_, yticklabels=le.classes_,
            cmap='Blues')
plt.xlabel('True label')
plt.ylabel('Predicted label');  
```

The confusion matrix confirms that the model is good at predicting when an individual has no sleeping disorder. However it seems to be quite ineffective at predicting the presence of sleep apnea and insomnia given when given sleep quality and stress level. We can also see that the model often gets confused between insomnia cases and no disorder cases. This implies that perhaps more data is required or that different features could be used to create a more effective model. Perhaps sleep duration and age may be better idicators of sleep disorders.

```python
# Shows labels
print(le.classes_)

# Shows posterior probabilities of the last 8 points
yprob = model.predict_proba(X_test)
yprob[-8:].round(2)
```

This array gives the posterior probabilities of the last 8 points in the test set. The first column corresponds to the probability of each individual having insomnia, the second column corresponds to the probability of the individual not having any sleeping disorder and the third column corresponds to the probability of the individual having sleep apnea.
