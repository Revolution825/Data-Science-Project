Standard Imports

```python
%matplotlib inline
import pandas as pd
import matplotlib. pyplot as plt
plt.style.use('seaborn-v0_8-whitegrid')
import numpy as np
```

Loading the Data in:

```python
dataset = pd.read_csv("Sleep_Health_and_Lifestyle_Dataset.csv")
```

Simple Linear Regressions

This first linear regression aims to show the relationship between stress level and sleep quality. It is expected that higher stress levels will result in lower sleep quality.

```python
from sklearn.linear_model import LinearRegression

x=dataset[["Stress Level"]]
y=dataset[["Quality of Sleep"]]

model = LinearRegression(fit_intercept=True)

model.fit(x, y)

xfit = pd.DataFrame(np.linspace(0, 10, 100).reshape(-1, 1), columns=["Stress Level"])
yfit = model.predict(xfit)

plt.ylim(0,10)
plt.scatter(x, y, s=10)
plt.plot(xfit, yfit, color="red")
plt.xlabel("Stress Level")
plt.ylabel("Quality of Sleep")
plt.title("Linear Regression: Stress Level vs Sleep Quality")
```

```python
slope = model.coef_[0].item()
r_squared = model.score(x, y)
r_value = np.sign(slope) * np.sqrt(r_squared)

print("Slope : " + str(round(slope, 2)))
print("R Squared Value : " + str(round(r_squared, 2)))
print("R Value : " + str(round(r_value, 2)))
```

The R value of -0.9 tells us that there is a significant negative correalation between Quality of Sleep (the dependent varible) and Stress Level (the independent variable). An R Squared value of 0.81 indicates that 81% of the variation in Quality of Sleep can be explained by its relationship with Stress Level.

This second linear regression aims to show the relationship between Quality of Sleep and Physical Activity Level. It is expected that a higher Physical Activity Level will result in a higher sleep quality.

```python
x2=dataset[["Physical Activity Level"]]
y2=dataset[["Quality of Sleep"]]

model2 = LinearRegression(fit_intercept=True)

model2.fit(x2, y2)

xfit2 = pd.DataFrame(np.linspace(0, 100, 100).reshape(-1, 1), columns=["Physical Activity Level"])
yfit2 = model2.predict(xfit2)

plt.ylim(0,10)
plt.scatter(x2, y2, s=10)
plt.plot(xfit2, yfit2, color="red")
plt.xlabel("Physical Activity Level")
plt.ylabel("Quality of Sleep")
plt.title("Linear Regression: Physical Activity Level vs Sleep Quality")
```

```python
slope2 = model2.coef_[0].item()
r_squared2 = model2.score(x2, y2)
r_value2 = np.sign(slope2) * np.sqrt(r_squared2)

print("Slope : " + str(round(slope2, 2)))
print("R Squared Value : " + str(round(r_squared2, 2)))
print("R Value : " + str(round(r_value2, 2)))
```

The R value of 0.19 tells us that there is a slight positive correalation between Quality of Sleep (the dependent varible) and Physical Activity (the independent variable). An R Squared value of 0.04 indicates that 4% of the variation in Quality of Sleep can be explained by its relationship with the Physical Activity.

This third linear regression aims to show the relationship between Quality of Sleep and Age. It is expected that quality of sleep decreases as age increases

```python
x3=dataset[["Age"]]
y3=dataset[["Quality of Sleep"]]

model3 = LinearRegression(fit_intercept=True)

model3.fit(x3, y3)

xfit3 = pd.DataFrame(np.linspace(0, 100, 100).reshape(-1, 1), columns=["Age"])
yfit3 = model3.predict(xfit3)

plt.ylim(0,10)
plt.scatter(x3, y3, s=10)
plt.plot(xfit3, yfit3, color="red")
plt.xlabel("Age")
plt.ylabel("Quality of Sleep")
plt.title("Linear Regression: Age vs Sleep Quality")
```

```python
slope3 = model3.coef_[0].item()
r_squared3 = model3.score(x3, y3)
r_value3 = np.sign(slope3) * np.sqrt(r_squared3)

print("Slope : " + str(round(slope3, 2)))
print("R Squared Value : " + str(round(r_squared3, 2)))
print("R Value : " + str(round(r_value3, 2)))
```

The R value of 0.47 tells us there is a positive correalation between Quality of Sleep and Age. An R squared value of 0.22 tells us that 22% of the variation in Quality of Sleep can be explained by its relationship with Age.
